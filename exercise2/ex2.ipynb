```python
import pandas as pd
import numpy as np
from scipy.optimize import linprog 
from numpy.linalg import solve
import urllib.request
import os
import matplotlib.pyplot as plt
```

```python
power_g = np.array([15, 100, 32, 25, 70]) 
cost_g = np.array([75, 15, 0, 42, 10])

g = np.vstack((power_g, cost_g)).T
g = g[g[:,1].argsort()]
ng = power_g.shape[0]

power_d = 180

ids = [''] * (ng)

for i in range(ng):
    ids[i] = 'G{}'.format(i+1)

ids = np.array(ids)
```

```python
xg = np.cumsum(g[:,0])
xg = np.insert(xg, 0, 0)
yg = g[:,1]
yg = np.insert(yg, 0, 0)
xd = power_d

plt.figure(figsize=(15, 7))
plt.step(xg,yg, where = 'pre', label ='Supply')
plt.axvline(x=xd, color = 'red')
plt.legend()
```
